DISTRIBUTION PLOTS

```python
import seaborn as sns
import numpy as np
import sklearn 
import pandas as pd
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.dtypes
```

```python
sdf=df[['total_bill','tip','size']]
sdf.corr()
```

```python
sns.heatmap(sdf.corr())
```

```python
sns.jointplot(x='tip',y='total_bill',data=sdf, kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=sdf, kind='reg')
```

```python
sns.pairplot(sdf)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.displot(df['tip'])
```

```python
sns.displot(df['tip'],kde=False,bins=10)
```
